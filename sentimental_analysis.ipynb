```python
import os
```

```python

# Function to read words from a file and return them as a set
def read_words_from_file(filename):
    words = set()
    with open(filename, 'r', encoding='latin1') as file:
        for line in file:
            word = line.strip()
            words.add(word)
    return words
```

```python

# Define the folder path containing text files
folder_path = r'C:\Users\suraj\OneDrive\Desktop\Projects\internship_applications_project\web_scrapper_with_sentimental_analysis\output_txt_files'


# Load positive and negative words
positive_words = read_words_from_file(r'C:\Users\suraj\OneDrive\Desktop\Projects\internship_applications_project\web_scrapper_with_sentimental_analysis\MasterDictionary\positive-words.txt')
negative_words = read_words_from_file(r'C:\Users\suraj\OneDrive\Desktop\Projects\internship_applications_project\web_scrapper_with_sentimental_analysis\MasterDictionary\negative-words.txt')
```

```python

# Initialize variables to store total positive and negative scores
total_positive_score = 0
total_negative_score = 0
num_files_processed = 0

# Iterate over each file in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.txt'):
        # Read the content of the file
        with open(os.path.join(folder_path, filename), 'r', encoding='utf-8') as file:
            text = file.read().lower()  # Convert text to lowercase for case-insensitive matching

        # Calculate positive and negative scores
        positive_score = sum(1 for word in text.split() if word in positive_words)
        negative_score = sum(1 for word in text.split() if word in negative_words)
        
        word_count = len(text.split())

        print(f"File: {filename}")
        print(f"Positive Score: {positive_score}")
        print(f"Negative Score: {negative_score}")
        print(f"Word Count: {word_count}\n")

        # Aggregate the scores
        total_positive_score += positive_score
        total_negative_score += negative_score
        # Increment the count of processed files
        num_files_processed += 1

# Print total positive and negative scores
print("number of files processed:",num_files_processed)
print("Total Positive Score:", total_positive_score)
print("Total Negative Score:", total_negative_score)
print("Total number of words are:", word_count)
```
